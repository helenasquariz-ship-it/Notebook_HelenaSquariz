## üßÆ Simulando um conjunto de dados para classifica√ß√£o  
Vamos gerar dados sint√©ticos representando um produto digital que classifica se um usu√°rio √© **propenso a cancelar uma assinatura**.


````python

import numpy as np
import pandas as pd
from sklearn.datasets import make_classification

# Simula√ß√£o dos dados
import numpy as np
import pandas as pd

np.random.seed(42)  # Para reprodutibilidade

# N√∫mero de registros
n = 500

# Vari√°veis num√©ricas
idade = np.random.randint(18, 70, n)
tempo_uso = np.random.exponential(scale=24, size=n).clip(1, 120).round(1)  # em meses
interacoes = np.random.poisson(lam=tempo_uso / 5).clip(0, None)
satisfacao = np.random.randint(1, 5 + 1, n)  # escala 1 a 5
reclamacoes = np.random.poisson(lam=(6 - satisfacao) / 2).clip(0, None)
gasto_mensal = np.random.normal(300, 120, n).clip(20, 1500).round(2)

# Vari√°veis categ√≥ricas
genero = np.random.choice(['Feminino', 'Masculino', 'N√£o-bin√°rio', 'Prefiro n√£o informar'],
                          n, p=[0.48, 0.48, 0.02, 0.02])

raca = np.random.choice([
    'Branca', 'Preta', 'Parda', 'Amarela', 'Ind√≠gena', 'Prefiro n√£o informar'
], n, p=[0.42, 0.10, 0.45, 0.01, 0.01, 0.01])

area_formacao = np.random.choice([
    'Humanas', 'Exatas', 'Biol√≥gicas', 'Neg√≥cios', 'Tecnologia', 'Artes'
], n)

# Construindo o DataFrame
df = pd.DataFrame({
    'idade': idade,
    'tempo_uso': tempo_uso,
    'interacoes': interacoes,
    'satisfacao': satisfacao,
    'reclamacoes': reclamacoes,
    'gasto_mensal': gasto_mensal,
    'genero': genero,
    'raca': raca,
    'area_formacao': area_formacao
})

# ----- CRIA√á√ÉO DA VARI√ÅVEL ALVO (CHURN) -----
# Calcula score de cancelamento
score_churn = (
    (5 - df['satisfacao']) * 0.4 +
    (df['reclamacoes']) * 0.3 +
    (1 / (df['interacoes'] + 1)) * 0.2 +
    (1 / (df['tempo_uso'] + 1)) * 0.1
)

# Normaliza score para (0,1)
score_churn = (score_churn - score_churn.min()) / (score_churn.max() - score_churn.min())

# Probabilidade aumenta conforme o score
df['churn'] = np.random.binomial(1, score_churn)

df.head()

````

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# -------------------
# 1. Selecionando features e alvo
# -------------------
X = df.drop('churn', axis=1)
y = df['churn']

# -------------------
# 2. Identificando tipos de vari√°veis
# -------------------
numeric_features = ['idade', 'tempo_uso', 'interacoes', 'satisfacao', 'reclamacoes', 'gasto_mensal']
categorical_features = ['genero', 'raca', 'area_formacao']

# -------------------
# 3. Pr√©-processamento
# -------------------
numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(drop='first', handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)



# Ajustar e transformar X
X_prepared = preprocessor.fit_transform(X)

feature_names = preprocessor.get_feature_names_out()
print(feature_names)

# -------------------
# 4. Divis√£o treino / teste
# -------------------
X_train, X_test, y_train, y_test = train_test_split(X_prepared, y, test_size=0.2, random_state=42)

print("Formato de X ap√≥s transforma√ß√£o:", X_prepared.shape)
print("Formato de X_train:", X_train.shape)
print("Formato de y_train:", y_train.shape)

pd.DataFrame(X_train)
````

üí° **Perguntas:**

Quando avaliamos um modelo de classifica√ß√£o (como previs√£o de churn), utilizamos algumas m√©tricas importantes que se baseiam na **matriz de confus√£o**:

|                            | Previsto N√£o Churn | Previsto Churn |
|----------------------------|-------------------|----------------|
| **Real N√£o Churn**         | TN (Verdadeiro Negativo) | FP (Falso Positivo) |
| **Real Churn**             | FN (Falso Negativo) | TP (Verdadeiro Positivo) |

- **TP**: Modelo acertou ao prever churn.
- **FP**: Modelo acusou churn, mas o cliente n√£o churnaria.
- **TN**: Modelo acertou ao prever n√£o churn.
- **FN**: Modelo n√£o detectou um churn verdadeiro.

---

### **Acur√°cia (Accuracy)**

> Mede o percentual de previs√µes corretas.

\[
\text{Acur√°cia} = \frac{TP + TN}{TP + TN + FP + FN}
\]

‚úÖ Boa quando as classes s√£o **balanceadas**  
‚ö†Ô∏è Pode ser enganosa em casos **desbalanceados**

---

### **Precis√£o (Precision)**

> Quando o modelo diz que o cliente vai churnar, qual √© a chance de estar certo?

\[
\text{Precision} = \frac{TP}{TP + FP}
\]

- Alta precis√£o ‚Üí poucos falsos alarmes.
- √ötil quando o custo de agir sem necessidade √© alto (ex: dar desconto para quem n√£o churnaria).

---

### **Recall (Sensibilidade)**

> Entre os clientes que realmente churnaram, quantos o modelo conseguiu detectar?

\[
\text{Recall} = \frac{TP}{TP + FN}
\]

- Alto recall ‚Üí o modelo n√£o deixa churn passar despercebido.
- √ötil quando perder o cliente custa caro.

---

### **F1-Score**

> Equil√≠brio entre **Precision** e **Recall**.

\[
\text{F1} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
\]

- Ideal quando as classes s√£o desbalanceadas.
- Mostra o compromisso entre precis√£o e sensibilidade.

---

### **Resumo**

| M√©trica     | Pergunta-chave | Ideal para... | Aten√ß√£o |
|-------------|----------------|---------------|---------|
| **Acur√°cia** | "Quantos acertos no total?" | Dados balanceados | Pode enganar quando h√° classe rara |
| **Precis√£o** | "Quando o modelo prev√™ churn, ele est√° certo?" | Minimizar falsos positivos | Pode perder churn real |
| **Recall**   | "O modelo encontra a maioria dos churns?" | Minimizar falsos negativos | Pode gerar falsos alarmes |
| **F1-score** | "Qual o equil√≠brio entre precis√£o e recall?" | Classes desbalanceadas | N√£o diferencia custos dos erros |



##  üì¶ Diferen√ßa entre ROC e AUC

A **ROC (Receiver Operating Characteristic Curve)** √© uma **curva** usada para avaliar o desempenho de modelos de classifica√ß√£o bin√°ria.  
Ela mostra como o modelo se comporta ao variar o limiar (threshold) que define a classe prevista.

A curva ROC √© constru√≠da usando:

- **Eixo X:** *False Positive Rate (FPR)* ‚Äî taxa de falsos positivos  
- **Eixo Y:** *True Positive Rate (TPR)* ‚Äî taxa de verdadeiros positivos (tamb√©m chamada de **Recall**)

Cada ponto da curva representa um threshold diferente para transformar a probabilidade prevista em classe (`0` ou `1`).

---

### AUC (Area Under the Curve)

A **AUC** √© um **valor num√©rico** que representa a **√°rea sob a curva ROC**.

- Variando de **0 a 1**
- Quanto mais pr√≥ximo de **1**, melhor o modelo
- Um modelo **aleat√≥rio** tem AUC ‚âà **0.5**
- Um modelo **perfeito** tem AUC = **1.0**

---

### Interpreta√ß√£o Intuitiva da AUC

A AUC pode ser interpretada como:

> A probabilidade de o modelo atribuir uma probabilidade maior ao exemplo positivo do que ao exemplo negativo.

Exemplo:
- Se **AUC = 0.85**, significa que **85% das vezes** o modelo colocar√° um caso positivo acima de um negativo no ranking de probabilidades.

---

### Resumo

| Conceito | Tipo | O que representa? | Valor |
|---------|------|------------------|-------|
| **ROC** | Curva | Rela√ß√£o TPR vs FPR ao variar o threshold | N√£o possui valor √∫nico |
| **AUC** | N√∫mero | √Årea sob a curva ROC | De 0 a 1 (quanto maior, melhor) |

---

### Quando usar ROC/AUC?

ROC e AUC s√£o especialmente √∫teis quando:
- Os dados s√£o **desbalanceados**
- Queremos avaliar o **ranking** das previs√µes, e n√£o apenas a classifica√ß√£o final
- Queremos comparar modelos independentemente do threshold escolhido

---

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
````

````python
## Fa√ßa outros gr√°ficos na etapa de data understanding
# Gr√°fico 1: Distribui√ß√£o de Churn por N√≠vel de Satisfa√ß√£o
plt.figure(figsize=(8, 6))
# Calcula a taxa de churn (m√©dia da vari√°vel 'churn') para cada n√≠vel de 'satisfacao'
churn_by_satisfaction = df.groupby('satisfacao')['churn'].mean().reset_index()

sns.barplot(x='satisfacao', y='churn', hue='satisfacao', data=churn_by_satisfaction, palette="Reds_d", legend=False)
plt.title('Taxa de Churn por N√≠vel de Satisfa√ß√£o do Usu√°rio')
plt.xlabel('N√≠vel de Satisfa√ß√£o (1=Muito Baixa, 5=Muito Alta)')
plt.ylabel('Taxa M√©dia de Churn (Propor√ß√£o)')
plt.grid(axis='y', linestyle='--')
plt.show()
````

````python
##Teste outros modelos na etapa de modelagem
# Defini√ß√£o dos modelos
models = {
    'RandomForest': RandomForestClassifier(random_state=42, n_jobs=-1),
    'GradientBoosting': GradientBoostingClassifier(random_state=42)
}

print("--- Treinamento e Avalia√ß√£o de Modelos Alternativos ---")

# Loop para treinar e avaliar cada modelo
for name, model in models.items():
    # Cria√ß√£o do pipeline para o modelo
    pipeline_model = Pipeline(steps=[('preprocessor', preprocessor),
                                     ('classifier', model)])

    # Treinamento
    pipeline_model.fit(X_train, y_train)

    # Previs√£o e Avalia√ß√£o
    y_pred = pipeline_model.predict(X_test)

    print(f"\nModelo: {name}")
    print("\nRelat√≥rio de Classifica√ß√£o (Threshold 0.5):")
    # O classification_report √© ideal pois exibe Precision, Recall e F1-Score
    print(classification_report(y_test, y_pred))
````

````python
## Utilize o grid search para encontrar os melhores hiperparametros do modelo
# Defini√ß√£o do Pipeline com o classificador
pipeline_rf = Pipeline(steps=[('preprocessor', preprocessor),
                              ('classifier', RandomForestClassifier(random_state=42))])

# Par√¢metros a serem testados
param_grid = {
    'classifier__n_estimators': [100, 250],  # N√∫mero de √°rvores
    'classifier__max_depth': [5, 10, None],  # Profundidade m√°xima da √°rvore
    'classifier__min_samples_split': [2, 5], # M√≠nimo de amostras para split
}

# Inicializa√ß√£o do Grid Search
grid_search = GridSearchCV(
    estimator=pipeline_rf,
    param_grid=param_grid,
    cv=3,  # Cross-Validation com 3 folds
    scoring='f1', # Otimizando o F1-Score
    n_jobs=-1,
    verbose=1
)

print("--- Iniciando Grid Search para RandomForestClassifier ---")

# Execu√ß√£o do Grid Search nos dados de treino
grid_search.fit(X_train, y_train)

# Exibe os resultados
print(f"\nMelhores par√¢metros encontrados: {grid_search.best_params_}")
print(f"Melhor F1-Score (m√©dia CV): {grid_search.best_score_:.4f}")

# Avalia√ß√£o do melhor modelo no conjunto de teste
best_model_rf = grid_search.best_estimator_
y_pred_tuned = best_model_rf.predict(X_test)

print("\nRelat√≥rio de Classifica√ß√£o do Modelo Otimizado (Conjunto de Teste):")
print(classification_report(y_test, y_pred_tuned))
````

````python
## Altere o threshold de classifica√ß√£o de um modelo e analise qual o melhor th
# Treinamento do modelo base (Logistic Regression)
pipeline_lr = Pipeline(steps=[('preprocessor', preprocessor),
                              ('classifier', LogisticRegression(random_state=42))])
pipeline_lr.fit(X_train, y_train)

# Obter probabilidades de previs√£o para a classe '1' (Churn)
y_proba = pipeline_lr.predict_proba(X_test)[:, 1]

# Lista de thresholds para testar, de 0.10 a 0.90
thresholds = np.arange(0.10, 0.91, 0.05).round(2)
metrics_list = []

print("--- An√°lise de M√©tricas Variando o Threshold de Classifica√ß√£o ---")

for th in thresholds:
    # Classifica√ß√£o com o novo threshold
    y_pred_th = (y_proba >= th).astype(int)

    # C√°lculo das m√©tricas (usando zero_division=0 para evitar warnings)
    precision = precision_score(y_test, y_pred_th, zero_division=0)
    recall = recall_score(y_test, y_pred_th, zero_division=0)
    f1 = f1_score(y_test, y_pred_th, zero_division=0)

    metrics_list.append({
        'Threshold': th,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1
    })

metrics_df = pd.DataFrame(metrics_list)
print(metrics_df)

# Plot para visualiza√ß√£o do trade-off
plt.figure(figsize=(10, 6))
plt.plot(metrics_df['Threshold'], metrics_df['Precision'], label='Precision', marker='o')
plt.plot(metrics_df['Threshold'], metrics_df['Recall'], label='Recall (Sensibilidade)', marker='o')
plt.plot(metrics_df['Threshold'], metrics_df['F1-Score'], label='F1-Score', marker='o')

plt.axvline(x=0.5, color='gray', linestyle='--', label='Threshold Padr√£o (0.5)')
plt.title('Trade-off entre Precision e Recall por Threshold')
plt.xlabel('Threshold de Classifica√ß√£o')
plt.ylabel('Valor da M√©trica')
plt.legend()
plt.grid(True, linestyle='--')
plt.show()

# Conclus√£o
best_f1_row = metrics_df.loc[metrics_df['F1-Score'].idxmax()]
print(f"\nO F1-Score m√°ximo ({best_f1_row['F1-Score']:.4f}) ocorre no Threshold de {best_f1_row['Threshold']}.")
print("\nAn√°lise Estrat√©gica:")
print("Se o seu foco principal √© a **reten√ß√£o de clientes** (minimizar Falsos Negativos), voc√™ deve escolher um threshold **abaixo** do ponto de F1 m√°ximo, como 0.30 ou 0.35, para obter um **Recall** mais alto, garantindo que voc√™ identifique a maior parte dos clientes que realmente ir√£o cancelar.")
````
